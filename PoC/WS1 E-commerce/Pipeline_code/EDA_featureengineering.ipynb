```python
import pandas as pd
import numpy as np
import os
import json
import pprint
from IPython.display import display

# Cấu hình Pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)

print("Các thư viện đã sẵn sàng.")

# ---------------------------------------------------------------------------
# PHẦN 1: TẢI DỮ LIỆU SẠCH (TỪ NOTEBOOK 1)
# ---------------------------------------------------------------------------
CLEAN_DATA_FILE = 'olist_master_table_CLEAN_v1.csv' # File này là ĐẦU RA của Notebook 1

print(f"Đang tải Bảng Dữ liệu SẠCH: {CLEAN_DATA_FILE}...")
try:
    # Tải bảng SẠCH
    df = pd.read_csv(CLEAN_DATA_FILE)

    # Quan trọng: Phải parse lại datetime sau khi tải CSV
    datetime_cols = [
        'order_purchase_timestamp', 'order_approved_at',
        'order_delivered_carrier_date', 'order_delivered_customer_date',
        'order_estimated_delivery_date', 'shipping_limit_date',
        'review_creation_date', 'review_answer_timestamp'
    ]
    for col in datetime_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors='coerce')

    print(f"✓ Tải thành công. Shape: {df.shape}")
    print("Kiểm tra 5 dòng đầu (dữ liệu sạch):")
    display(df.head())

except FileNotFoundError:
    print(f"LỖI: Không tìm thấy file {CLEAN_DATA_FILE}.")
    print("Vui lòng đảm bảo bạn đã chạy '01_Merge_and_Clean.ipynb' thành công trước.")
except Exception as e:
    print(f"LỖI KHÁC KHI TẢI FILE: {e}")
```

# PHẦN 2 - TẠO ĐẶC TRƯNG (FEATURE ENGINEERING)

```python
# ---------------------------------------------------------------------------
# PHẦN 2: TẠO ĐẶC TRƯNG (FEATURE ENGINEERING)
# ---------------------------------------------------------------------------
print("\n" + "="*50)
print("BẮT ĐẦU TẠO ĐẶC TRƯNG (FEATURE ENGINEERING)...")
print("="*50)

# --- 2.1: Đặc trưng Thời gian (Time-based Features) ---
# Reasoning: Tạo ra các đặc trưng liên quan đến tốc độ và độ tin cậy của việc giao hàng.
print("Đang tạo đặc trưng Thời gian...")

# 1. Thời gian giao hàng thực tế (ngày)
df['delivery_time_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.total_seconds() / (24 * 60 * 60)
df['estimated_delivery_time_days'] = (df['order_estimated_delivery_date'] - df['order_purchase_timestamp']).dt.total_seconds() / (24 * 60 * 60)
df['delivery_vs_estimated_days'] = (df['order_estimated_delivery_date'] - df['order_delivered_customer_date']).dt.total_seconds() / (24 * 60 * 60)
df['order_processing_time_days'] = (df['order_delivered_carrier_date'] - df['order_purchase_timestamp']).dt.total_seconds() / (24 * 60 * 60)
# 5. clinical

df['purchase_year'] = df['order_purchase_timestamp'].dt.year
df['purchase_month'] = df['order_purchase_timestamp'].dt.month
df['purchase_day_of_week'] = df['order_purchase_timestamp'].dt.dayofweek # 0=Thứ 2, 6=Chủ Nhật
df['purchase_hour'] = df['order_purchase_timestamp'].dt.hour

# Bạn cũng có thể tạo đặc trưng nhị phân cho cuối tuần
df['is_weekend'] = df['purchase_day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

print("✓ Đã tạo các đặc trưng cyclical (month, day_of_week, hour).")

# Xử lý NaT (đơn hàng chưa giao):
# Reasoning: Các đơn chưa giao (NaT) sẽ tạo ra 'delivery_time_days' = NaN.
# Chúng ta dùng giá trị lính canh -999 (logic từ 'Audit and Clean.ipynb')
time_cols_to_fill = ['delivery_time_days', 'delivery_vs_estimated_days', 'order_processing_time_days']
df[time_cols_to_fill] = df[time_cols_to_fill].fillna(-999)
print("✓ Đã tạo các đặc trưng thời gian (delivery_time, processing_time...).")


# --- 2.2: Đặc trưng Địa lý (Geo-based Features) ---
# Reasoning: Tính khoảng cách vật lý.
print("Đang tạo đặc trưng Địa lý (Khoảng cách Haversine)...")

def haversine_np(lat1, lon1, lat2, lon2):
    R = 6371  # Bán kính Trái Đất (km)
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    distance = R * c
    return distance

# Áp dụng hàm (vectorized)
# Vì đã Clean ở Notebook 1, các cột lat/lng này không còn NaN
df['dist_cust_seller_km'] = haversine_np(
    df['customer_lat'],
    df['customer_lng'],
    df['seller_lat'],
    df['seller_lng']
)
print("✓ Đã tạo đặc trưng 'dist_cust_seller_km'.")


# --- 2.3: Đặc trưng Tài chính & Sản phẩm (Financial & Product) ---
# Reasoning: Tạo ra các tỷ lệ và thông số vật lý của sản phẩm.
print("Đang tạo đặc trưng Tài chính và Sản phẩm...")

# 1. Tỷ lệ Phí vận chuyển / Giá sản phẩm
#    Xử lý lỗi Chia cho 0 (vì 'price' đã được clean > 0)
df['freight_ratio'] = df['freight_value'] / df['price']
# Thay thế giá trị vô cực (nếu có) bằng 0
df['freight_ratio'] = df['freight_ratio'].replace([np.inf, -np.inf], 0)

# 2. Thể tích sản phẩm
df['product_volume_cm3'] = \
    (df['product_length_cm'] * \
     df['product_height_cm'] * \
     df['product_width_cm'])

print("✓ Đã tạo đặc trưng 'freight_ratio' và 'product_volume_cm3'.")

print("Đang điền khuyết (impute) các đặc trưng sản phẩm bị thiếu...")
product_numeric_cols = [
    'product_name_lenght', 'product_description_lenght',
    'product_photos_qty', 'product_weight_g', 'product_length_cm',
    'product_height_cm', 'product_width_cm', 'product_volume_cm3'
]

for col in product_numeric_cols:
    if col in df.columns:
        df[col] = df[col].fillna(0) # Điền 0 cho các sản phẩm bị thiếu

print("✓ Đã điền 0 cho các đặc trưng sản phẩm bị thiếu (NaN).")

print("\n" + "="*50)
print("✓✓✓ TẠO ĐẶC TRƯNG HOÀN TẤT ✓✓✓")
print(f"Shape của Bảng Đặc trưng Cuối cùng: {df.shape}")
print("="*50)

# --- Hiển thị kết quả ---
print("Kiểm tra 5 dòng đầu của Bảng Đặc trưng Cuối cùng:")
display(df.head())
```

# validation

```python
# ---------------------------------------------------------------------------
# PHẦN 3: THẨM ĐỊNH CUỐI CÙNG (FINAL AUDIT)
# ---------------------------------------------------------------------------
print("\n" + "="*80)
print("BẮT ĐẦU THẨM ĐỊNH TOÀN DIỆN (COMPREHENSIVE VALIDATION)...")
print("="*80)

# Dán toàn bộ hàm 'comprehensive_validation' từ notebook 'Audit and Clean.ipynb'
# (Đây là hàm mà bạn đã sử dụng)
def comprehensive_validation(df, verbose=True):
    validation_results = {}

    # 3.1: Thông tin cơ bản
    if verbose: print("\n--- 3.1 Thông tin cơ bản DataFrame ---")
    validation_results['shape'] = df.shape
    validation_results['total_rows'] = df.shape[0]
    validation_results['total_cols'] = df.shape[1]
    validation_results['memory_mb'] = round(df.memory_usage(deep=True).sum() / 1024 ** 2, 2)
    if verbose:
        print(f"✓ Shape: {validation_results['shape']}")
        print(f"✓ Memory: {validation_results['memory_mb']} MB")

    # 3.2: Kiểm tra Missing Values
    if verbose: print("\n--- 3.2 Kiểm tra Missing Values ---")
    total_cells = df.shape[0] * df.shape[1]
    total_missing = df.isna().sum().sum()
    missing_pct = round(total_missing * 100.0 / total_cells, 2)
    validation_results['total_missing_values'] = total_missing
    validation_results['missing_pct'] = missing_pct
    validation_results['cols_with_missing'] = df.isna().any().sum()
    if verbose:
        print(f"✓ Total missing values: {total_missing:,} ({missing_pct}%)")
        print(f"✓ Columns with missing: {validation_results['cols_with_missing']}/{df.shape[1]}")
        top_missing = df.isna().sum().sort_values(ascending=False).head(5)
        top_missing = top_missing[top_missing > 0]
        if not top_missing.empty:
            print("\n  Top 5 cột thiếu nhiều nhất:")
            for col, count in top_missing.items():
                pct = round(count * 100.0 / df.shape[0], 2)
                print(f"    - {col}: {count:,} ({pct}%)")

    # 3.3: Kiểm tra Duplicates
    if verbose: print("\n--- 3.3 Kiểm tra Duplicates ---")
    dup_rows = df.duplicated().sum()
    validation_results['duplicate_rows'] = dup_rows
    if verbose:
        if dup_rows > 0: print(f"⚠ Duplicate rows: {dup_rows:,}")
        else: print(f"✓ No duplicate rows found")

    # 3.4: Kiểm tra Key Integrity (Uniqueness)
    if verbose: print("\n--- 3.5 Kiểm tra Key Integrity ---")
    key_stats = {}
    important_keys = ['order_id', 'customer_id', 'product_id', 'seller_id']
    for key in important_keys:
        if key in df.columns:
            unique_count = df[key].nunique()
            key_stats[key] = {'unique': unique_count}
            if verbose: print(f"✓ {key}: {unique_count:,} unique values")
    validation_results['key_integrity'] = key_stats

    # 3.5: Kiểm tra Business Logic Constraints
    if verbose: print("\n--- 3.6 Kiểm tra Business Logic Constraints ---")
    violations = {}
    if 'review_score' in df.columns:
        invalid_reviews = df[(df['review_score'] < 0) | (df['review_score'] > 5)]
        violations['invalid_review_score'] = len(invalid_reviews)
    if 'price' in df.columns: violations['negative_price'] = (df['price'] <= 0).sum()
    if 'freight_value' in df.columns: violations['negative_freight'] = (df['freight_value'] < 0).sum()
    if 'delivery_time_days' in df.columns:
        invalid_delivery = (df['delivery_time_days'] < -999).sum()
        violations['invalid_delivery_time'] = invalid_delivery
    if verbose: print(f"✓ Total Business Logic Violations: {sum(violations.values())}")
    validation_results['business_logic_violations'] = violations

    # 3.6: Kiểm tra Sentinel Values
    if verbose: print("\n--- 3.8 Kiểm tra Sentinel Values ---")
    sentinel_stats = {}
    if 'review_score' in df.columns:
        sentinel_stats['review_score_zero'] = (df['review_score'] == 0).sum()
    if 'product_category_name_english' in df.columns:
        sentinel_stats['product_category_unknown'] = (df['product_category_name_english'] == 'unknown').sum()
    if 'delivery_time_days' in df.columns:
        sentinel_stats['delivery_not_completed'] = (df['delivery_time_days'] == -999).sum()
    validation_results['sentinel_values'] = sentinel_stats
    if verbose: print(f"✓ Sentinel values stats calculated (review=0, category=unknown, delivery=-999)")

    # 3.7: Validation Score (Overall Quality Score)
    score = 100.0
    score -= min(missing_pct * 5, 25)  # Max -25 điểm
    score -= min((dup_rows * 100.0 / df.shape[0]) * 5, 15) # Max -15 điểm
    total_violations = sum(violations.values())
    violation_pct = total_violations * 100.0 / df.shape[0] if df.shape[0] > 0 else 0
    score -= min(violation_pct * 10, 20) # Max -20 điểm
    validation_results['quality_score'] = round(max(score, 0), 2)

    if verbose:
        print(f"\n--- 3.10 Overall Data Quality Score ---")
        print(f"🎯 Quality Score: {validation_results['quality_score']}/100")
        if validation_results['quality_score'] >= 90: print("✅ EXCELLENT - Data is production ready!")
        elif validation_results['quality_score'] >= 75: print("✓ GOOD - Minor issues, acceptable.")
        else: print("⚠ POOR - Significant data quality issues found.")

    return validation_results

# Chạy validation trên Bảng Đặc trưng cuối cùng (df)
validation_report = comprehensive_validation(df, verbose=True)
```
